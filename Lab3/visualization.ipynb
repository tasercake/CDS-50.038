```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline
from scipy.stats import linregress
import numpy as np

sns.set()
mpl_params = mpl.rcParams
```

```python
dataframe = pd.read_csv("scottish_hills.csv")
dataframe.head()
```

```python
sorted_hills = dataframe.sort_values(by=['Height'], ascending=False)
sorted_hills.head()
```

```python
x = dataframe.Height
y = dataframe.Latitude
z = dataframe.Longitude
```

# Scatterplots

```python
stats = linregress(x,y)
m, b = stats.slope, stats.intercept
plt.figure(figsize=(8, 4))
plt.scatter(x, y)
plt.plot(x, m * x + b, color='red')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
colors = np.random.rand(len(y))
plt.scatter(y, z, s=x - 800, c=colors, alpha=0.5)
plt.show()
```

# Histograms

```python
plt.figure(figsize=(8, 4))
plt.hist(x, bins=25, density=True)
plt.show()
```

```python
shifted_x = x - 100
fig, ax = plt.subplots(1)
plt.figure(figsize=(8, 4))
ax.hist(x, bins=25, density=True, histtype='stepfilled', alpha=0.7, label='Height')
ax.hist(shifted_x, bins=25, density=True, histtype='stepfilled', alpha=0.7, label='Heigh - 100')
ax.legend(prop={'size': 10})
ax.set_xlabel('Height')
ax.set_ylabel('Normalised Distribution')
plt.show()
```

# Bar Plot

```python
labels = ['EU', 'US', 'AS']
meateaters = [122, 135, 80]
vegetarians = [40, 42, 23]
# dataframe = pd.DataFrame({'labels': labels, 'meateaters': meateaters, 'vegetarians': vegetarians})
indices = np.arange(len(labels))
plt.figure(figsize=(8, 4))
fig, ax = plt.subplots()
bar_width = 0.35
rects1 = ax.bar(indices, meateaters, bar_width, color='r', label='Meat Eaters')
rects2 = ax.bar(indices + bar_width, vegetarians, bar_width, color='b', label='Vegetarians')
ax.set_xlabel('Continent')
ax.set_ylabel('CO2 Emissions')
ax.set_title('CO2 Emissions per continent, from diet')
ax.set_xticks(indices + bar_width / 2)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
```

# Line Plot

```python
def fahrenheit2celcius(temp):
    return (temp - 32) * 5 / 9
```

```python
temperature = [100, 102, 106, 105, 90, 85, 89, 100, 102, 103, 108, 100, 102, 106, 105, 90, 85, 89, 100, 102, 103, 108]
fig, ax_f = plt.subplots()
ax_c = ax_f.twinx()

ax_f.plot(temperature)
ax_f.set_xlim(0, 24)
y1, y2 = ax_f.get_ylim()
ax_c.set_ylim(fahrenheit2celcius(y1), fahrenheit2celcius(y2))
ax_c.figure.canvas.draw()

ax_f.set_title("Two Scales: Fahrenheit and Celcius")
ax_f.set_ylabel("Fahrenheit")
ax_c.set_ylabel("Celcius")
ax_f.set_xlabel("Hour of day")
plt.show()
```

# Assignment

```python
url = "https://raw.githubusercontent.com/plotly/datasets/master/school_earnings.csv"
df = pd.read_csv(url)
print(df.head())
print(df.describe())
```

```python
sns.reset_defaults()
mpl.rcParams.update(mpl_params)
plt.style.use("dark_background")
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df['Women'], bins=12, alpha=0.7, density=True, label='Women')
ax.hist(df['Men'], bins=18, alpha=0.7, density=True, label='Men')
ax.set_xlabel('Salary')
ax.set_ylabel('Normalised Distribution')
plt.legend()
plt.show()
plt.style.use("default")
```

```python
plt.figure(figsize=(10, 7))
melted = pd.melt(df, id_vars=['School', 'Gap'], var_name='Sex', value_name='Salary').sort_values('Gap')
sns.set()
ax = sns.barplot(x='School', y='Salary', hue='Sex', data=melted)
ax.set_xticklabels(ax.get_xticklabels(), rotation=60)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(x='Sex', y='Salary', data=melted)
plt.show()
```
